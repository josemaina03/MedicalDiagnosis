```python
class DrugPrescription:
    def __init__(self):
        self.prescribed_drugs = []

    def add_prescription(self, drug_name):
        if drug_name not in self.prescribed_drugs:
            self.prescribed_drugs.append(drug_name)
            print(f"{drug_name} has been prescribed.")
        else:
            print(f"{drug_name} is already in the prescription.")

    def remove_prescription(self, drug_name):
        if drug_name in self.prescribed_drugs:
            self.prescribed_drugs.remove(drug_name)
            print(f"{drug_name} has been removed from the prescription.")
        else:
            print(f"{drug_name} is not in the prescription.")

    def get_prescription(self):
        if self.prescribed_drugs:
            print("Prescribed drugs:")
            for drug in self.prescribed_drugs:
                print(drug)
        else:
            print("No drugs prescribed.")

    def clear_prescription(self):
        self.prescribed_drugs = []
        print("Prescription has been cleared.")

    def get_unique_drugs(self):
        unique_drugs = set(self.prescribed_drugs)
        if unique_drugs:
            print("Unique drugs prescribed:")
            for drug in unique_drugs:
                print(drug)
        else:
            print("No unique drugs prescribed.")

    def get_prescription_count(self):
        prescription_count = len(self.prescribed_drugs)
        print(f"Number of drugs prescribed: {prescription_count}")


# Example usage
prescription = DrugPrescription()
prescription.add_prescription("Aspirin")
prescription.add_prescription("Ibuprofen")
prescription.add_prescription("Aspirin")  # Trying to add a duplicate prescription

prescription.get_prescription()
prescription.get_unique_drugs()
prescription.get_prescription_count()

prescription.remove_prescription("Paracetamol")  # Trying to remove a non-prescribed drug
prescription.remove_prescription("Aspirin")
prescription.get_prescription()
prescription.get_unique_drugs()
prescription.get_prescription_count()

prescription.clear_prescription()
prescription.get_prescription()
prescription.get_unique_drugs()
prescription.get_prescription_count()
```
